```python
import pandas as pd
from datetime import date
import numpy as np
```

```python
financials_annual = pd.read_excel("AAPL-fs-04292022-nohfe.xlsx").transpose()
```

```python
new_header = financials_annual.iloc[0]
financials_annual = financials_annual[1:]
financials_annual.columns = new_header
financials_annual = financials_annual.iloc[: , :-4]
financials_annual.drop("SEC Link", axis=1, inplace=True)
financials_annual.head(2)
```

```python
financials_annual.replace("- -", np.nan, inplace=True)

#records from only 2012 onwards
financials_annual = financials_annual.iloc[27:, :]

financials_annual.dropna(axis=1, inplace=True)
```

```python
pd.set_option('display.max_columns', 1000) 
financials_annual.head()
```

```python
cols = financials_annual.columns
len(cols)

index = financials_annual.index
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

financials_annual = scaler.fit_transform(financials_annual)
financials_annual = pd.DataFrame(financials_annual, columns=cols)
```

```python
financials_annual
```

```python
financials_annual.to_csv("aapl_financials_annual_2012_to_21.csv", index=False)
financials_annual = pd.read_csv("aapl_financials_annual_2012_to_21.csv")
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=10)
pca.fit(financials_annual)
financials_annual = pca.transform(financials_annual)
```

```python
pca.explained_variance_ratio_
```

```python
financials_annual = pd.DataFrame(financials_annual)
financials_annual
```
